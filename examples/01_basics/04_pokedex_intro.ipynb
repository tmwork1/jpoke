# POKEDEX の紹介

`jpoke.data.POKEDEX` を使って、ポケモンの図鑑データ（持てる特性・覚えられる技など）を
調べる方法を扱う。

バトルを実行せず、静的なデータだけで完結する一番手軽な調べ物の例。

Google Colabで開いた場合は、まず次のセルで `jpoke` をインストールする
（ローカルで `pip install jpoke` 済みなら不要）。

```python
%pip install -q jpoke
```

```python
from __future__ import annotations

from jpoke import Pokemon
from jpoke.data import POKEDEX
```

`Pokemon` を作ると、そのポケモンの図鑑データに `mon.data` でアクセスできる
（詳細は docs/api/README.md の POKEDEX（図鑑データ）節を参照）。持てる特性は
`mon.data.abilities` で確認できる（セル末尾の式はNotebook上でそのまま表示される）。

```python
mon = Pokemon("ピカチュウ")
mon.data.abilities
```

覚えられる技は `mon.data.learnset` で確認できる（`list[MoveName]`、五十音順にソート済み）。

```python
mon.data.learnset[:10]  # 最初の10個だけ表示
```

同じデータは、`Pokemon` を作らなくても `POKEDEX[name]` で直接調べられる
（`mon.data` はこの `POKEDEX[name]` そのもの）。

```python
dex = POKEDEX["ピカチュウ"]
print(f"ピカチュウが持てる特性: {dex.abilities}")
print(f"mon.dataと同一のオブジェクトか: {dex is mon.data}")
```

注意: add_pokemon(ability_name=...) を省略した場合の挙動は
docs/api/README.md の Pokemon コンストラクタ節を参照
（POKEDEX の情報を見て手動で指定する必要がある）
