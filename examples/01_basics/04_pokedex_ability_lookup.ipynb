# POKEDEX を使って、ポケモンが持てる特性・覚えられる技を確認する

POKEDEX を使って、ポケモンが持てる特性・覚えられる技を確認する方法を扱う。

バトルを実行せず、`jpoke.data.POKEDEX` という静的なデータだけで完結する
一番手軽な調べ物の例。

Google Colabで開いた場合は、まず次のセルで `jpoke` をインストールする
（ローカルで `pip install jpoke` 済みなら不要）。

```python
%pip install -q jpoke
```

```python
from __future__ import annotations

from jpoke.data import POKEDEX
```

ポケモンごとに持てる特性は `POKEDEX[name].abilities` で確認できる
（詳細は docs/api/README.md の POKEDEX（図鑑データ）節を参照）。

```python
for name in ("ライチュウ", "フシギバナ", "ピカチュウ"):
    print(f"{name}が持てる特性: {POKEDEX[name].abilities}")
```

ポケモンが覚えられる技の集合は `POKEDEX[name].learnset` で確認できる
（詳細は docs/api/README.md の POKEDEX（図鑑データ）節を参照。ここでは件数と
`sorted()` した先頭数件だけ表示する）。

```python
for name in ("ライチュウ", "フシギバナ", "ピカチュウ"):
    learnset = POKEDEX[name].learnset
    print(f"{name}が覚えられる技: {len(learnset)}種類（例: {sorted(learnset)[:5]}）")
```

注意: add_pokemon(ability_name=...) を省略した場合の挙動は
docs/api/README.md の Pokemon コンストラクタ節を参照
（POKEDEX の情報を見て手動で指定する必要がある）
