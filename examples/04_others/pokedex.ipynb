`Pokemon` を作ると、そのポケモンの図鑑データに `mon.data` でアクセスできる。

```python
from jpoke import Pokemon

mon = Pokemon("ピカチュウ")
dex = mon.data

print(f"特性一覧: {dex.abilities}")
print(f"覚える技: {dex.learnset[:10]}") # 最初の10個だけ表示
print(f"対応レギュレーション: {dex.regulations}")
```

`POKEDEX[name]` で直接調べることも可能。

```python
from jpoke.data import POKEDEX

dex = POKEDEX["ピカチュウ"]

print(f"特性一覧: {dex.abilities}")
print(f"覚える技: {dex.learnset[:10]}") # 最初の10個だけ表示
print(f"対応レギュレーション: {dex.regulations}")
```

あるレギュレーションに該当するポケモン・アイテムの一覧を取得したい場合は、
`get_pokemon_by_regulation()` / `get_items_by_regulation()` を使う（戻り値は五十音順）。

```python
from jpoke import get_pokemon_by_regulation, get_items_by_regulation

pokemon_list = get_pokemon_by_regulation("M-A")
item_list = get_items_by_regulation("M-A")

print(f"M-Aで使用可能なポケモン数: {len(pokemon_list)}")
print(f"M-Aで使用可能なアイテム数: {len(item_list)}")
print(f"先頭5件（ポケモン）: {pokemon_list[:5]}")
print(f"先頭5件（アイテム）: {item_list[:5]}")
```
