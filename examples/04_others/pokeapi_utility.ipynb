`get_pokeapi_url()` で和名からPokeAPIのエンドポイントURLを取得できる。ポケモン・アイテムどちらも対応する。

```python
from jpoke import get_pokeapi_url

print(get_pokeapi_url("ピカチュウ"))                   # ポケモン（category省略時は"pokemon"）
print(get_pokeapi_url("たべのこし", category="item"))  # アイテム
```

`get_pokemon_image_url()` / `get_item_image_url()` は画像（スプライト）のURLを返す。ポケモンは `image_type` で公式アートワーク・ドット絵・色違いなど種類を指定できる。

```python
from jpoke import get_pokemon_image_url, get_item_image_url

print(get_pokemon_image_url("ピカチュウ"))  # 既定は公式アートワーク
print(get_pokemon_image_url("ピカチュウ", image_type="showdown-front-shiny"))
print(get_item_image_url("たべのこし"))
```

URLではなく画像ファイル自体が欲しい場合は `download_pokemon_image()` / `download_item_image()` を使う。指定したパスへ画像をダウンロードして保存し、保存先の `Path` を返す（保存先ディレクトリが存在しなければ自動作成される）。

```python
import tempfile
from pathlib import Path

from jpoke import download_pokemon_image, download_item_image

dest_dir = Path(tempfile.gettempdir()) / "jpoke_images"
pikachu_path = download_pokemon_image("ピカチュウ", dest_dir / "pikachu.png")
leftovers_path = download_item_image("たべのこし", dest_dir / "leftovers.png")

print(f"{pikachu_path} ({pikachu_path.stat().st_size} bytes)")
print(f"{leftovers_path} ({leftovers_path.stat().st_size} bytes)")
```

和名がPokeAPI側に見つからない場合、上記いずれの関数も `PokeApiResolveError` を送出する。

```python
from jpoke.exceptions import PokeApiResolveError

try:
    get_pokeapi_url("存在しないポケモン")
except PokeApiResolveError as e:
    print(f"解決できませんでした: {e}")
```
