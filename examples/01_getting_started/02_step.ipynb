```python
from jpoke import Battle
from jpoke.players import RandomPlayer
```

```python
player1 = RandomPlayer("Team A")
player1.add_pokemon(
    "ピカチュウ", ability="せいでんき", item="いのちのたま", moves=["かみなり"],
)
player1.add_pokemon(
    "ヒトカゲ", ability="もうか", item="オボンのみ", moves=["かえんほうしゃ"],
)
player1.add_pokemon(
    "フシギダネ", ability="しんりょく", item="くろいヘドロ", moves=["ギガドレイン"],
)

player2 = RandomPlayer("Team B")
player2.add_pokemon(
    "ゼニガメ", ability="げきりゅう", item="オボンのみ", moves=["なみのり"],
)
player2.add_pokemon(
    "コラッタ", ability="はりきり", item="ちからのハチマキ", moves=["すてみタックル"],
)
player2.add_pokemon(
    "ピッピ", ability="マジックガード", item="オボンのみ", moves=["ムーンフォース"],
)
```

```python
battle = Battle(player1, player2, seed=1)

# バトル開始処理。ポケモンの選出はここで行われる。
battle.start()

# 決着がつくかターン上限に達するまで手動でstep()する
while battle.can_continue(max_turns=30):
    battle.step()
    battle.print_logs()
    print("-" * 50)

winner = battle.winner
print(f"勝者: {winner.username if winner else '引き分け（ターン上限）'}")
```
