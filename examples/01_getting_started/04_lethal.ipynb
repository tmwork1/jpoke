```python
from __future__ import annotations

from jpoke import Battle, Player
```

```python
attack_player = Player("Attack Player")
attack_player.add_pokemon(
    "ガブリアス", evs={"atk": 32},
)

defence_player = Player("Defence Player")
defence_player.add_pokemon(
    "カイリュー", ability_name="マルチスケイル"
)

battle = Battle(attack_player, defence_player)
battle.start()

attacker = battle.get_active(attack_player)
defender = battle.get_active(defence_player)
```

```python
# Pokemon.show() で実数値・性格・特性・持ち物・テラスタイプ・技構成をまとめて確認できる
attacker.show()
defender.show()
```

```python
results = battle.calc_lethal(
    attacker=attacker,
    moves="ドラゴンテール",
    critical=False,
    move_secondary=False,
)

result = results[-1]
print(f"ダメージ: {result.min_damage}~{result.max_damage}")
print(f"{result.n_attack}発 致死率 {result.lethal_probability:.2%}")
```
