# 木探索する AI を作る

TreeSearchPlayer は木探索でコマンドを選ぶ Player の基底クラス。

ここでは evaluate() をオーバーライドして葉ノード（盤面）の評価方法をカスタマイズする

```python
from jpoke import Battle
from jpoke.players import RandomPlayer, TreeSearchPlayer
```

```python
class KOFocusedPlayer(TreeSearchPlayer):
    """相手を瀕死にできる手を優先するAI"""

    def evaluate(self, battle: Battle) -> float:
        # 既定の評価（残りHP割合の差）に、相手の瀕死数をボーナスとして加える
        base = super().evaluate(battle)
        opponent_team = battle.get_team(battle.opponent(self))
        n_fainted = sum(1 for mon in opponent_team if mon.fainted)
        return base + n_fainted
```

```python
ai_player = KOFocusedPlayer("TreeSearch", max_plies=1, max_nodes=50)
ai_player.add_pokemon("カビゴン", move_names=["のしかかり", "じしん"])
ai_player.add_pokemon("ギャラドス", move_names=["たきのぼり"])

random_player = RandomPlayer("Random")
random_player.add_pokemon("ピカチュウ", move_names=["でんきショック"])
random_player.add_pokemon("ライチュウ", move_names=["10まんボルト"])

battle = Battle(ai_player, random_player, seed=1)
battle.play_out(max_turns=10)
battle.print_logs("all")
```
