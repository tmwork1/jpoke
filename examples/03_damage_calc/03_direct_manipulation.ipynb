```python
from __future__ import annotations
from jpoke import Battle, Player
```

```python
attack_player = Player("Attack Player")
attack_player.add_pokemon("ガブリアス")

defence_player = Player("Defence Player")
defence_player.add_pokemon("カイリュー")

battle = Battle(attack_player, defence_player)
battle.start()

attacker = battle.get_active(attack_player)
defender = battle.get_active(defence_player)
```

```python
# HP
defender.hp = defender.max_hp // 2
```

```python
# 能力ランク
attacker.boosts['atk'] = 2
```

```python
# 状態異常
battle.set_ailment(defender, "どく")
print(f"{defender.name}の状態異常: {defender.status}") # defender.ailment.name でも同じ結果が得られる
```

```python
# 揮発状態
battle.set_volatile(defender, "やどりぎのタネ")
```

```python
# 共通の場の状態
battle.set_weather("あめ")
battle.set_terrain("エレキフィールド")
```

```python
# プレイヤーの場の状態
battle.get_side(attack_player).apply("リフレクター", 5)
```
