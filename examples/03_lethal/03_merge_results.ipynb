# 複数の致死率計算結果を合成する

LethalHitResultどうしを足すことで加算ダメージを計算できる

```python
%pip install -q jpoke
```

```python
from jpoke import Battle, Player
```

```python
attack_player = Player("Attack Player")
attack_player.add_pokemon("ガブリアス")

defence_player = Player("Defence Player")
defence_player.add_pokemon("カイリュー")

battle = Battle(attack_player, defence_player)
battle.start()
```

```python
# 一つ目の技の計算
results1 = battle.calc_lethal(
    attacker=battle.get_active(attack_player),
    moves=["ドラゴンテール"],
)
result = results1[-1]
print(f"技: {result.move.name}")
print(f"攻撃回数: {result.attack_count}")
print(f"ダメージ: {result.min_damage}~{result.max_damage}")
print(f"HP分布: {result.hp_counter}")
print(f"致死率: {result.lethal_probability:.2%}")
```

```python
# 二つ目の技の計算
results2 = battle.calc_lethal(
    attacker=battle.get_active(attack_player),
    moves=["いわなだれ"],
)
result = results2[-1]
print(f"技: {result.move.name}")
print(f"攻撃回数: {result.attack_count}")
print(f"ダメージ: {result.min_damage}~{result.max_damage}")
print(f"HP分布: {result.hp_counter}")
print(f"致死率: {result.lethal_probability:.2%}")
```

```python
# 二つの技の結果を合成
result = results1[0] + results2[0]
print(f"技: {result.move.name}")
print(f"攻撃回数: {result.attack_count}")
print(f"ダメージ: {result.min_damage}~{result.max_damage}")
print(f"HP分布: {result.hp_counter}")
print(f"致死率: {result.lethal_probability:.2%}")
```
