```python
from jpoke import Battle, Player
```

```python
attack_player = Player("Attack Player")
attack_player.add_pokemon("ガブリアス")

defence_player = Player("Defence Player")
defence_player.add_pokemon("カイリュー")

battle = Battle(attack_player, defence_player)
battle.start()

attacker = battle.get_active(attack_player)
defender = battle.get_active(defence_player)
```

```python
# HP
defender.hp = defender.max_hp // 2

# 能力ランク
attacker.boosts['atk'] = 1

# 状態異常
battle.set_ailment(defender, "どく")

# 揮発状態
battle.set_volatile(defender, "やどりぎのタネ")

# 共通の場の状態
battle.set_weather("あめ")
battle.set_terrain("エレキフィールド")

# プレイヤーの場の状態
battle.get_side(defence_player).apply("リフレクター", 5)

attacker.show()
defender.show()
```

```python
results = battle.calc_lethal(
    attacker=attacker,
    moves="アイアンヘッド",
    critical=False,
    move_secondary=False,
)

print(f"技: {results[0].move.name}")
for result in results:
    print("-"*50)
    print(f"攻撃回数: {result.attack_count}")
    print(f"ヒット回数: {result.hit_count}")
    print(f"ダメージ: {result.min_damage}~{result.max_damage}")
    print(f"HP分布: {result.hp_counter}")
    print(f"致死率: {result.lethal_probability:.2%}")
```
